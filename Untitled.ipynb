```python
import pandas as pd
from sklearn import tree

dogs = pd.read_csv("https://stepik.org/media/attachments/course/4852/cats.csv")
dogs = dogs.drop('Unnamed: 0', axis=1)

dogs_X = dogs.iloc[:, :3]
dogs_y = dogs.iloc[:, 3]

dogs_clf = tree.DecisionTreeClassifier(criterion='entropy', random_state=0)
dogs_clf.fit(dogs_X, dogs_y)

tree.plot_tree(dogs_clf, feature_names=dogs_X.columns)
```

```python
# Подключаем (импортируем) специализированные модули Pandas и Numpy, 
# не являющиеся частью стандартной библиотеки Python.
import pandas as pd
import numpy as np

# Считываем файл c данными о кошках и собаках и сохраняем его в переменную dogs_cats.
dogs_cats = pd.read_csv('https://stepik.org/media/attachments/course/4852/cats.csv')

# Сохраним в отдельные переменные все значения х и у.
X = dogs_cats[['Шерстист', 'Гавкает', 'Лазает по деревьям']]
y = dogs_cats['Вид']

# Определяем функцию по расчету энтропии.
# На вход подается список значений микросостояний системы для рассматриваемого макросостояния.
def entropy(n_list):
    e = 0
    sum = 0
    # Вычисление суммы значений микросостояний системы.
    for i in range(len(n_list)):
        sum = sum + n_list[i]
    # Вычисление значения энтропии для рассматриваемого макросостояния системы, 
    # как суммы энтропий микросостояний (i).
    for i in range(len(n_list)):
        e = e - (n_list[i] / sum) * np.log2(n_list[i] / sum)
    return e

# Цикл по каждой особенности системы - столбцам DataFrame "Шерстист", "Гавкает", "Лазает по деревьям".
for feature in X.columns:
    # Описание параметров каждого макросостояния системы.
    print('Особенность:', feature)
    print(dogs_cats.loc[:, [feature, 'Вид']])
    p = X[feature].unique()
    print('Уникальные значения "', feature, '": ', p, sep='')
    print('Общее количество значений "', feature, '" = ', X[feature].count(), sep='')
    print()
    for i in range(len(p)):
        print('Уникальные значения "Вид" при "', feature, ': ', p[i], '" = ', y[X[feature] == p[i]].unique(), 
              sep='')
        print('Общее количество значений "Вид" при "', feature, ': ', p[i], '" = ', y[X[feature] == p[i]].count(),
              sep='')
        arg_to_def = list()
        # Описание параметров каждого микросостояния системы.
        for j in range(len(y[X[feature] == p[i]].unique())):
            print('Количество значений "Вид: ', y[X[feature] == p[i]].unique()[j], '" = ',
                  y[(X[feature] == p[i]) & (y == y[X[feature] == p[i]].unique()[j])].count(), sep='')
            # Формирование списка значений микросостояний системы для рассматриваемого макросостояния.
            arg_to_def.append(y[(X[feature] == p[i]) & (y == y[X[feature] == p[i]].unique()[j])].count())
        print('Энтропия при разделении по особенности "', feature, '" в группах, где "', feature, '" равно ', 
              p[i], '" = ', '%.2f' %entropy(arg_to_def), sep='')
        print()
```
