```python
from sklearn import tree
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
iris = sns.load_dataset("iris")
#setosar = iris.loc[iris.species == "setosar"]
#virginica = iris.loc[iris.species == "virginica"]
g = sns.JointGrid(x="sepal_width", y="petal_length", data=iris)
#sns.kdeplot(setosa.sepal_width, setosa.sepal_length, cmap="Reds",
            #shade=True, shade_lowest=False, ax=g.ax_joint)
#sns.kdeplot(virginica.sepal_width, virginica.sepal_length, cmap="Blues",
            #shade=True, shade_lowest=False, ax=g.ax_joint)
sns.kdeplot(iris.sepal_width, iris.sepal_length, cmap="Reds",
            shade=True, shade_lowest=False, ax=g.ax_joint)
sns.kdeplot(iris.petal_width, iris.petal_length, cmap="Blues",
            shade=True, shade_lowest=False, ax=g.ax_joint)
sns.distplot(iris.sepal_width, kde=True, hist=False, color="r", ax=g.ax_marg_x)
sns.distplot(iris.petal_width, kde=True, hist=False, color="b", ax=g.ax_marg_x)
sns.distplot(iris.petal_length, kde=True, hist=False, color="r", ax=g.ax_marg_y, vertical=True)
sns.distplot(iris.sepal_length, kde=True, hist=False, color="b", ax=g.ax_marg_y, vertical=True)
plt.show()
```
